```python
from transformers import pipeline
import torch

# Initialize the sentiment analysis pipeline
sentiment_pipeline = pipeline('sentiment-analysis')

def analyze_sentiment(text):
    """
    Analyze the sentiment of the given text.
    :param text: The input text to analyze.
    :return: The sentiment label and score.
    """
    result = sentiment_pipeline(text)
    return result

def main():
    print("Real-time Sentiment Analysis")
    print("Type 'exit' to quit.")

    while True:
        # Get input from user
        user_input = input("Enter text: ")

        if user_input.lower() == 'exit':
            print("Exiting...")
            break

        # Analyze sentiment
        sentiment_result = analyze_sentiment(user_input)
        print(f"Sentiment Analysis Result: {sentiment_result}")

if __name__ == "__main__":
    main()
```

```python
from transformers import pipeline
import torch

# Initialize the sentiment analysis pipeline with a model that supports three classes
sentiment_pipeline = pipeline('sentiment-analysis', model='cardiffnlp/twitter-roberta-base-sentiment')

def analyze_sentiment(text):
    """
    Analyze the sentiment of the given text.
    :param text: The input text to analyze.
    :return: The sentiment label and score.
    """
    result = sentiment_pipeline(text)
    return result

def main():
    print("Real-time Sentiment Analysis")
    print("Type 'exit' to quit.")

    while True:
        # Get input from user
        user_input = input("Enter text: ")

        if user_input.lower() == 'exit':
            print("Exiting...")
            break

        # Analyze sentiment
        sentiment_result = analyze_sentiment(user_input)
        for res in sentiment_result:
            print(f"Sentiment: {res['label']}, Score: {res['score']}")

if __name__ == "__main__":
    main()
```
